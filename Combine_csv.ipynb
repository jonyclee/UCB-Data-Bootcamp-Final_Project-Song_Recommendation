```python
# Dependencies
import pandas as pd
import score_converter
```

```python
# Read billboards csv
songs_df = pd.read_csv('./csv/billboards_cleaned.csv')
songs_df.head()
```

```python
# Read lyrics csv
lyric_df = pd.read_csv('./csv/lyrics.csv')
lyric_df.head()
```

```python
# Read sentiment analysis scores csv
scores_df = pd.read_csv('./csv/scores.csv')
scores_df.head()
```

```python
# Rescale sentiment analysis
old = scores_df['Compound_Score'].values
new = score_converter.convert(old)
new_score = []
for i in new:
    temp = i.round(2).item()
    new_score.append(temp)
```

```python
# Read songs info csv
info_df = pd.read_csv('./csv/songs_info.csv')
info_df
```

```python
# Read top words csv
words_df = pd.read_csv('./csv/top_words.csv')
words_df
```

```python
# Take wanted columns from dataframes
artists = songs_df['Artist']
titles = songs_df['Title']
lyrics = lyric_df['Lyric']
scores = scores_df['Compound_Score']
genres = info_df['Genre'].values
moods = info_df['Mood'].values
tempos = info_df['Tempo'].values
top_words = words_df['Top_Three_Words']
```

```python
# Create new df
df = pd.DataFrame({'Artist': artists,
                  'Title': titles,
                  'Lyric': lyrics,
                  'Compound_Score': scores,
                  'Genre': genres,
                  'Mood': moods,
                  'Tempo': tempos,
                  'Top_Word': top_words,
                  'New_Score': new_score})
df.head()
```

```python
# Export df to csv
df.to_csv('./csv/main.csv')
```
