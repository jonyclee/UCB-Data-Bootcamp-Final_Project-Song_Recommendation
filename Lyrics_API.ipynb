```python
# Import Dependencies
from PyLyrics import *
import pandas as pd
```

```python
# Read csv with songs and artist
songs_df = pd.read_csv('./csv/billboards.csv')
songs_df
```

```python
# List of songs and artist
songs = songs_df['Title']
artist_temp = songs_df['Artist']
artist = []
for i in range(0, len(artist_temp)):
    temp = artist_temp[i].split('Featuring')
    temp = temp[0].split('&')
    temp = temp[0].split(',')
    singer = temp[0].strip()
    artist.append(str(singer))
```

```python
# Gets lyrics using PyLyrics. Saves the index of unavailable songs
lyrics = []
drop_index = []
for j in range(0,len(songs)):
    try:
        lyric = PyLyrics.getLyrics(artist[j], songs[j])
        lyrics.append(lyric)
    except:
        drop_index.append(j) 
```

```python
# Drops the song from dataframe if lyrics is not available
songs_new_df = songs_df
for i in range(len(drop_index)-1,0,-1):
    ind = drop_index[i]
    songs_new_df = songs_new_df.drop(songs_new_df.index[ind])
ind = drop_index[0]
songs_new_df = songs_new_df.drop(songs_new_df.index[ind])
```

```python
songs_new_df.head()
```

```python
# Converts new songs df to csv
songs_new_df.to_csv('./csv/billboards_cleaned.csv')
```

```python
# Saves list of lyrics into a dataframe
lyric_df = pd.DataFrame({'Lyric': lyrics})
```

```python
# Converts lyrcs df to csv
lyric_df.to_csv('./csv/lyrics.csv')
```
